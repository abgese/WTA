```python
import pandas as pd
import numpy as np
from math import ceil
```

```python
def RatingMatrix( data ) :
	m = len( data.groupby( 'User' ).aggregate( 'count' ) )
	n = len( data.groupby( 'Movie' ).aggregate( 'count' ) )
	R = np.zeros( ( m , n ) )
	for i in data.index :
		u = data.User[ i ]
		m = data.Movie[ i ]
		R[ u - 1 ][ m - 1 ]=data.Rating[ i ]
        return R
```

```python
len_attack = 100
```

```python
data = pd.read_csv('u.tsv',delimiter='\t')
```

```python
R = RatingMatrix( data )
```
